# We have created a new data frame for the means,we can remove the word '_mean' from the columns as it is meaningless now

```python
cancer_means.columns
```

```python
#remove _mean from columns and save to new series

new_labels = []

for col in cancer_means.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])
    else:
        new_labels.append(col)
        
new_labels    
```

```python
# assign new labels to columns in dataframe
cancer_means.columns = new_labels
```

```python
#Notice '_mean' has been removed
cancer_means.head()
```

```python
# save this for later
# If I did not use index = False then jupyter would automatically add an new index column to the saved table so we would have two index columns
cancer_means.to_csv('cancer_data_means.csv', index = False)
```
